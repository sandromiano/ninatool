```python
from ninatool.internal.elements import L, J
from ninatool.internal.structures import loop
import numpy as np
import matplotlib.pyplot as plt
```

```python
#maximum order of expansion, order = 1 means linear order
order = 4

#left branch elements
J0 = J(1, order = order, name = 'J0')
L0 = L(.1, order = order, name = 'L0')
#right branch elements
J1 = J(.5, order = order, name = 'J1')
L1 = L(.1, order = order, name = 'L1')

dcsquid = loop(
    left_branch = [J0, L0], 
    right_branch = [J1, L1], 
    name = 'dcsquid', 
    stray_inductance = False)
```

```python
#sets the phase of the free JJ
dcsquid.free_phi = np.linspace(-1,1,101) * 2 * np.pi
```

```python
#plots second order expansion coefficient (u2) VS flux
plt.plot(dcsquid.flux, dcsquid.adm[0])
plt.xlabel('dcsquid.flux')
plt.ylabel('dcsquid.u2')
```

```python
#plots third order expansion coefficient (u3) VS flux
plt.plot(dcsquid.flux, dcsquid.adm[1])
plt.xlabel('dcsquid.flux')
plt.ylabel('dcsquid.u3')
```

```python
#plots second order expansion coefficient (u2) VS flux, with changes in J1 critical current

#this syntax updates the critical current of J1 and automatically updates the dcsquid.
#can be applied to linear inductance as well.
J1.ic = 0.4
plt.plot(dcsquid.flux, dcsquid.adm[0], label = 'J1.ic=' + str(J1.ic))
J1.ic = 0.6
plt.plot(dcsquid.flux, dcsquid.adm[0], label = 'J1.ic=' + str(J1.ic))
plt.legend()
plt.xlabel('dcsquid.flux')
plt.ylabel('dcsquid.u2')
```
